```python
BIGDF = []
artists = ["JT", "NS"]
curr = '/Users/chelseaparlett/Desktop/Desktop/Crash Course/unique_lyrs/'
for artist in artists:
    newwd = curr +  artist
    os.chdir(newwd)
#     print(os.getcwd())
    data = []
    f = lambda x: x.lower().replace(",", "").replace("[", "").replace("]", "").replace("(","").replace(")","").replace(".", "").replace("?", "").replace("!", "")
    songList = glob.glob('*.txt')
#     print(songList)
    for song in songList:
        unique_items = len(set(f(w) for w in open(song).read().split()))
        name = song.split('.')[0]
        data.append([name, unique_items])


    dataFrame = pd.DataFrame(data)
    dataFrame.columns = ['name', 'unique words']
    print dataFrame
    y = dataFrame['unique words'].tolist()
    BIGDF.append(y)
    os.chdir(curr)

bp = plt.boxplot(BIGDF, positions = [1, 2], widths = 0.6)
plt.xticks([1,2],['Justin', 'NSYNC'])
plt.show()

for i in [0,1]:
    print(i)
    print("mean", np.mean(BIGDF[i]))
    print('median', np.percentile(BIGDF[i], 50))
    print('Q1', np.percentile(BIGDF[i], 25))
    print('Q3', np.percentile(BIGDF[i], 75))
    
```
